## create function to extracted all articles title and content

```python
articles_dic = {
    'URL_ID' : [],
    'URL' : [],
    'Title' : [],
    'Content' : []
}
def return_article_title_content(url) : 
    soup = BeautifulSoup(requests.get(url, headers = headers).text, 'lxml')
    return soup.find('h1', class_  = 'entry-title').text, soup.find('div', class_ = 'td-post-content').text

for i in range(df_output.shape[0]) : 
    
    # each row of df_output dataFarame
    row = df_output.iloc[i,:]
    
    # row['URL'] => url of each article
    # return_article_text return as each Article all content of url row['URL']
    title, content = return_article_title_content(row['URL'])
    
    # append data in articles_dic
    articles_dic['URL_ID'].append(row['URL_ID'])
    articles_dic['URL'].append(row['URL'])
    articles_dic['Title'].append(title)
    articles_dic['Content'].append(content)
    
    # to save each article_content as row['URL_ID'] name in txt form in Extracted articles folder
    with open(f"Extracted articles/{row['URL_ID']}.txt", 'w', encoding='utf-8') as f : 
        f.write(title + " " + content)
        
    # indicate particular article content downloaded
    print(f"{row['URL']} completed...")
    # break
```

### Cleaning using Stop Words Lists
#### The Stop Words Lists (found in the folder StopWords) are used to clean the text so that Sentiment Analysis can be performed by excluding the words found in Stop Words List.


```python
os.listdir(r'StopWords/')
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

#### all stopwords stored in stopwords_set python set

```python
stopwords_set = set()
for itm in os.listdir(r"StopWords/") :
    with open(f'StopWords/{itm}', 'r') as f : 
        stop = f.read()
        # print(stop)
        for wd in stop.split("\n") :
            first_wd = wd.split(' ')
            stopwords_set.add(first_wd[0].lower())
            print(first_wd[0].lower())
```

```python
stopwords_set.remove('')
stopwords_set
```

```python
positive_words_set = set()
with open(os.path.join(MasterDictionary_pt, 'positive-words.txt'), 'r') as f : 
    pws = f.read()
    # print(stop)
    for wd in pws.split('\n') :
        positive_words_set.add(wd.lower())
        print(wd.lower())
```

```python
import re
def clean_text(row) : 
    row = re.sub("\n", "", row).lower()
    return " ".join([itm.lemma_ for itm in nlp(row) if not (itm in stopwords_set or itm.is_punct or itm.is_stop or itm.is_space ) ])
```

```python
clean_text("My name is Rakesh Kumar Sahni. I am going to walk")
```

```python
df_articles['cleanContent'] = df_articles['Content'].apply(clean_text)
```

```python
df_articles.head()
```

```python
def positiveScore(row) : 
    ps = 0
    for wd in row['cleanContent'].split(" ") : 
        if wd in positive_words_set : 
            ps += 1
    return ps

def negativeScore(row) : 
    ns = 0
    for wd in row['cleanContent'].split(" ") : 
        if wd in negative_words_set : 
            ns += 1
    return ns

def polarityScore(row) : 
    return (row['ps'] - row['ns']) / ((row['ps'] + row['ns']) + 0.000001)

def subjectivityScore(row) : 
    total_words_after_cleaning = len(row['cleanContent'].split(" "))
    return (row['ps'] + row['ns'])/ ((total_words_after_cleaning) + 0.000001)
```

```python
df_articles['ps'] = df_articles.apply(positiveScore, axis = 1)
df_articles['ns'] = df_articles.apply(negativeScore, axis = 1)
df_articles['polar_s'] = df_articles.apply(polarityScore, axis = 1)
df_articles['subj_s'] = df_articles.apply(subjectivityScore, axis = 1)
```

```python
def avgSentLenght(row) : 
    row = re.sub("\n", "", row['Content']).lower()
    tot_words = 0
    tot_sents = 0
    doc_row = nlp(row)
    
    for itm in doc_row : 
        if not(itm.is_punct or itm.is_space) : 
            tot_words += 1
    
    for itm in doc_row.sents : 
        tot_sents += 1
    
    return tot_words / ( tot_sents + 0.0001)

def no_of_syllables(word) : 
    syllables_cnt = 0
    for ch in word : 
        if ch in 'aeiou' : 
            syllables_cnt += 1
            
    return syllables_cnt
    

def percentageComplexWords(row) : 
    row = re.sub("\n", "", row['Content']).lower()
    tot_words = 0
    tot_complex_words = 0
    
    doc_row = nlp(row)
    
    for itm in doc_row : 
        if not(itm.is_punct or itm.is_space) : 
            tot_words += 1
                    
            if no_of_syllables(itm.lemma_) > 2 : 
                tot_complex_words += 1
    
    return tot_words / ( tot_complex_words + 0.0001)

def fogIndex(row) : 
    return 0.4 * row['AVG SENTENCE LENGTH'] + row['PERCENTAGE OF COMPLEX WORDS']

def complexWordCount(row) :
    row = re.sub("\n", "", row['Content']).lower()
    tot_complex_words = 0
    
    doc_row = nlp(row)
    
    for itm in doc_row : 
        if not(itm.is_punct or itm.is_space) :         
            if no_of_syllables(itm.lemma_) > 2 : 
                tot_complex_words += 1
    
    return tot_complex_words

def wordCount(row) : 
    return len(row['cleanContent'].split(" "))

def syllableCountPerWord(row) : 
    row = re.sub("\n", "", row['Content']).lower()
    tot_syllable = 0
    tot_words =  0
    doc_row = nlp(row)
    
    for itm in doc_row : 
        if not(itm.is_punct or itm.is_space) :
            tot_words += 1
            tot_syllable += no_of_syllables(itm.lemma_) 
    
    return tot_syllable / (tot_words + 0.0001) 

def personalPronouns(row) : 
    row = re.sub("\n", "", row['Content']).lower()
    tot_personal_pronouns = 0
    pp_list = ['i', 'we', 'my', 'ours', 'us']
    doc_row = nlp(row)
    
    for itm in doc_row : 
        if not( itm.is_punct or itm.is_space) :         
            if itm.text in pp_list : 
                tot_personal_pronouns += 1
                
    return tot_personal_pronouns

def averageWordLength(row) : 
    row = re.sub("\n", "", row['Content']).lower()
    tot_syllable = 0
    
    doc_row = nlp(row)
    
    return sum([len(wd) for wd in doc_row]) / len(doc_row)
    
```

```python
df_articles['AVG SENTENCE LENGTH'] = df_articles.apply(avgSentLenght, axis = 1)
```

```python
df_articles['PERCENTAGE OF COMPLEX WORDS'] = df_articles.apply(percentageComplexWords, axis = 1)
```

```python
df_articles['FOG INDEX'] = df_articles.apply(fogIndex, axis = 1)
```

```python
df_articles['AVG NUMBER OF WORDS PER SENTENCE'] = df_articles['AVG SENTENCE LENGTH']
```

```python
df_articles['COMPLEX WORD COUNT'] = df_articles.apply(complexWordCount, axis = 1)
```

```python
df_articles['WORD COUNT'] = df_articles.apply(wordCount, axis = 1)
```

```python
df_articles['SYLLABLE PER WORD'] = df_articles.apply(syllableCountPerWord, axis = 1)
```

```python
df_articles['PERSONAL PRONOUNS'] = df_articles.apply(personalPronouns, axis = 1)
```

```python
df_articles['AVG WORD LENGTH'] = df_articles.apply(averageWordLength, axis = 1)
```
